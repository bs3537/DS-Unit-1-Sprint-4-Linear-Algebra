\begin{align}
G= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}

```
# Identity matrix for G

G = [(12, 11),
      (7, 10)]
G_matrix = np.matrix(G)

# Inverse of matrix G
Inv_G = np.linalg.inv(G_matrix)
Inv_G
```

```
#Identity matrix for G = G.inverse of G
IG = np.dot(G_matrix, Inv_G)
IG
# The second number in the first row and the first number in second row are almost equal to zero, so identity matrix for G is [(1, 0), (0, 1)]
```
